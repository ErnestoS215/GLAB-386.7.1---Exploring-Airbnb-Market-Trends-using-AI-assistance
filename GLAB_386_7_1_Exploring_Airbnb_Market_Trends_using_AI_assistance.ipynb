import pandas as pd

# Load the datasets
last_review = pd.read_csv("airbnb_last_review.tsv", sep="\t")
room_type = pd.read_excel("airbnb_room_type.xlsx")
price = pd.read_csv("airbnb_price.csv")

# Merge the datasets
merged_df = pd.merge(last_review, room_type, on="listing_id", how="inner")
merged_df = pd.merge(merged_df, price, on="listing_id", how="inner")

# Convert the price column to numeric
merged_df["price"] = (
    merged_df["price"]
    .replace(r"[\$,]", "", regex=True)
    .astype(float)
)

# Calculate the average listing price
average_price = merged_df["price"].mean()

# Print the result
print(f"Average Listing Price: ${average_price:.2f}")
Explanation
The three datasets are merged using the common key listing_id.
The price column is cleaned by removing the dollar sign ($) and any commas using .replace().
.astype(float) converts the cleaned prices into numeric values.
.mean() calculates the average listing price.
The result is printed with two decimal places using an f-string.

````python
import pandas as pd

# Load the datasets
last_review = pd.read_csv("airbnb_last_review.tsv", sep="\t")
room_type = pd.read_excel("airbnb_room_type.xlsx")
price = pd.read_csv("airbnb_price.csv")

# Merge the datasets
merged_df = pd.merge(last_review, room_type, on="listing_id", how="inner")
merged_df = pd.merge(merged_df, price, on="listing_id", how="inner")

# Clean the price column
merged_df["price"] = (
    merged_df["price"]
    .str.replace(" dollars", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Calculate the average listing price
average_price = merged_df["price"].mean()

# Print the result
print(f"Average Listing Price: ${average_price:.2f}")
````

Interpretation

The price data contains several potential outliers. The minimum price of $0 is unusual and may represent promotional listings, missing values, or data-entry errors. On the opposite end, the maximum price of $7,500 is much higher than the typical listing price.

Using the IQR method:

Q1 = $69
Q3 = $175
IQR = $106

The upper outlier threshold is:

Upper Bound = Q3 + 1.5 × IQR = 175 + (1.5 × 106) = $334

Any listing priced above $334 would be considered a potential outlier.

The lower outlier threshold is:

Lower Bound = Q1 − 1.5 × IQR = 69 − (1.5 × 106) = -$90

Since prices cannot be negative, there are no lower outliers by the IQR rule, although $0 listings should still be investigated.
